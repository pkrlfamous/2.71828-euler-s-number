```python
# ── Helper: run a module script and show its saved PNG ──────────────────────
from IPython.display import Image, display
import subprocess

def run_and_show(script_path, png_path):
    """Run a script (headless) then display the saved PNG."""
    result = subprocess.run(
        [sys.executable, script_path],
        capture_output=True, text=True, cwd=REPO
    )
    if result.stdout:
        print(result.stdout)
    if result.returncode != 0:
        print('STDERR:', result.stderr)
    display(Image(filename=os.path.join(REPO, png_path), width=900))

run_and_show('01_computing_e/limit_definition.py',
             '01_computing_e/outputs/limit_definition.png')
```

```python
# ── Live: verify Euler's identity ──────────────────────────────────────────
import cmath
result = cmath.exp(1j * math.pi) + 1
print(f'e^(iπ) + 1 = {result}  (should be 0+0j)')
print(f'|e^(iπ) + 1| = {abs(result):.2e}  (machine epsilon)')

# Unit circle: |e^(iθ)| = 1 for all θ
for theta in [0, math.pi/6, math.pi/4, math.pi/3, math.pi/2, math.pi]:
    z = cmath.exp(1j * theta)
    print(f'  θ={theta:.4f}: e^(iθ) = {z.real:+.4f}{z.imag:+.4f}i  |z|={abs(z):.6f}')
```

```python
# ── Live: Newton's cooling ─────────────────────────────────────────────────
T_env, T0, k = 20, 90, 0.1   # °C, °C, cooling constant
t = np.linspace(0, 60, 300)
T = T_env + (T0 - T_env) * np.exp(-k * t)

apply_euler_style()
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(t, T, color=COLORS['e_red'], lw=2.5, label=f'T(t) = {T_env} + ({T0}-{T_env})·e^(-{k}t)')
ax.axhline(T_env, color=COLORS['e_blue'], ls='--', lw=1.5, label=f'Ambient = {T_env}°C')
ax.set_xlabel('Time (min)'); ax.set_ylabel('Temperature (°C)')
ax.set_title("Newton's Law of Cooling — T(t) = Tₑ + (T₀-Tₑ)·e^(-kt)")
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.savefig('/tmp/cooling.png', dpi=120, bbox_inches='tight'); plt.close()
display(Image('/tmp/cooling.png', width=800))
print(f'Temperature after 30 min: {T_env + (T0-T_env)*math.exp(-k*30):.1f}°C')
```

```python
# ── Live: play with sigmoid and softmax ────────────────────────────────────
x = np.linspace(-6, 6, 200)

# Sigmoid
sig = sigmoid(x)
sig_deriv = sig * (1 - sig)   # σ'(x) = σ(x)(1-σ(x))

apply_euler_style()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
ax1.plot(x, sig, color=COLORS['e_gold'], lw=2.5, label='σ(x)')
ax1.plot(x, sig_deriv, color=COLORS['e_cyan'], lw=2.5, ls='--', label="σ'(x) = σ(1-σ)")
ax1.axhline(0.5, color='gray', ls=':', alpha=0.5)
ax1.set_title('Sigmoid & its derivative'); ax1.legend(); ax1.grid(True, alpha=0.3)

# Softmax with different temperatures
logits = [3.0, 1.0, 0.5]
temps  = [0.3, 1.0, 3.0]
x_pos  = np.arange(3)
for i, T in enumerate(temps):
    probs = softmax(logits, temperature=T)
    ax2.bar(x_pos + i*0.25, probs, 0.25, label=f'T={T}',
            color=[COLORS['e_red'], COLORS['e_gold'], COLORS['e_blue']][i], alpha=0.85)
ax2.set_xticks(x_pos + 0.25); ax2.set_xticklabels(['Logit 3.0', 'Logit 1.0', 'Logit 0.5'])
ax2.set_title('Softmax temperature: cold=sharp, hot=flat'); ax2.legend(); ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout(); plt.savefig('/tmp/ml_live.png', dpi=120, bbox_inches='tight'); plt.close()
display(Image('/tmp/ml_live.png', width=900))
```

```python
# ── Live: compare compounding frequencies ──────────────────────────────────
P, r, t = 1000, 0.08, 10   # principal, rate, years
print(f'${{P:,.0f}} invested at {{r:.0%}} for {{t}} years:')
print()
freqs = [('Annual', 1), ('Semi-annual', 2), ('Quarterly', 4),
         ('Monthly', 12), ('Daily', 365), ('Continuous', None)]
for name, n in freqs:
    if n is None:
        A = P * math.exp(r * t)
        label = 'Pe^(rt)'
    else:
        A = P * (1 + r/n)**(n*t)
        label = f'n={n}'
    print(f'  {name:<14} ({label:<10}): ${A:,.2f}  (gain ${A-P:,.2f})')
```

```python
# ── Live: Poisson distribution ─────────────────────────────────────────────
lam = 4   # average rate (change me!)
k_vals = range(0, 15)
probs  = [poisson_pmf(k, lam) for k in k_vals]

apply_euler_style()
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(k_vals, probs, color=COLORS['e_blue'], alpha=0.85, edgecolor=COLORS['e_cyan'], lw=1.2)
ax.axvline(lam, color=COLORS['e_red'], ls='--', lw=2, label=f'λ = {lam} (mean)')
ax.set_xlabel('k'); ax.set_ylabel('P(X=k)')
ax.set_title(fr'Poisson PMF: $P(k;\lambda) = \lambda^k e^{{-\lambda}} / k!$  with λ={lam}')
ax.legend(); ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout(); plt.savefig('/tmp/poisson.png', dpi=120, bbox_inches='tight'); plt.close()
display(Image('/tmp/poisson.png', width=800))
print(f'P(0; λ={lam}) = e^(-{lam}) = {math.exp(-lam):.6f}  (pure exponential connection)')
```
